```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import math
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
def readColumnsFromFile(filename):
    
    with open(filename) as infile:
        reader = csv.reader(infile, delimiter=",")
        ncol=len(next(reader)) # Read first line and count columns
        infile.seek(0) 
        
        all_cols = [[] for x in range(ncol)]
        
        for row in reader:
            row_data = [float(col.strip()) for col in row] 
            all_cols = [listo+[data] for listo, data in zip(all_cols, row_data)]
            
        return all_cols
```

```python
d_rates, bf_thicknesses, pop_sizes = readColumnsFromFile('optimal_detach_rates-thickness-popSize.txt')
bf_thicknesses = [x*5. for x in bf_thicknesses]
```

```python
def plot_detach_rate_diagnostics(det_rs, bf_thcks, pop_szs, pic_title):
    
    fig, ax = plt.subplots()
    ax2 = ax.twinx()
    fig.set_size_inches(6, 4)
    titleSize = 12
    labelSize = 12
    numbersize = 10
    legendsize = 12
    IDsize=12
    
    ax.plot(det_rs, bf_thcks, color='k')
    ax2.plot(det_rs, pop_szs, color='m')
    ax2.tick_params(axis='y', colors='m')
    
    ax.set_title('Varying detachment rate', fontsize=titleSize)
    ax.set_xlabel('detachment rate', fontsize=labelSize)
    ax.set_ylabel('biofilm thickness ($\mu$m)', fontsize=labelSize)
    ax2.set_ylabel('pop size', fontsize=labelSize)
    ax.xaxis.set_tick_params(labelsize=numbersize)
    ax.yaxis.set_tick_params(labelsize=numbersize)
    ax.set_xlim(0, 0.1)
    ax.set_ylim(bottom=0)
    ax2.set_ylim(bottom=0)
    plt.savefig(pic_title, bbox_inches='tight')
    plt.show()
    

plot_detach_rate_diagnostics(d_rates, bf_thicknesses, pop_sizes, 'thickness-popsize_vs_detachrate.pdf')
    
```
